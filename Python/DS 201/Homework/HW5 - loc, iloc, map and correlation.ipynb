In the next problem, we will use panda's loc function to select rows or columns.
The Pandas loc method enables you to select data from a Pandas DataFrame by label.
It allows you to “$\textbf{loc}$ate” data in a DataFrame. That’s where we get the name loc[]. We use it to locate data. loc has multiple input methods. You enter two inputs in loc function as  $loc[rows you want, columns you want]$.

A single scalar label:  df.loc$[ 1, columnname$] will select row 1(remember row 1 is actually the second row in python) and from that row, it picks the value under the column name. 

A list of labels:  df.loc$[[1,4,5], :]$ will select row 1, row 4, row 5 and all the columns.
Read more about loc function here.

Correlation coefficient: We have talked about correlation and what it means in our lecture. Correlation coefficent is a number that measures how two columns correlated. There are many kinds of correlation coefficients. We will use Pearson's Correlation coefficient. Given that the variables x and y have data points $x_1, x_2,..., x_n$ and $y_1, y_2,.., y_n$ then the correlation coefficient is defined by $r_{xy}=\dfrac{\displaystyle{\sum_{n=1}^n(x_i-\bar{x})(x_i-\bar{x})}}{\sigma_x\, \sigma_y}$.

$r_{xy}$ is the correlation coefficient. It's value ranges from -1 to 1. Correlation coefficent of -1 means highest possible negative correlation, 0 means no correlation and 1 means highest possible positive correlation.

$\bar{x}$ and $\bar{y}$ are means of x and y respectively.

$\sigma_x$ and $\sigma_y$ are standard deviation of x and y respectuvely. You do not have to worry about complicated looking formula. We will use pandas to find correlation. Let's first prepare the data.
